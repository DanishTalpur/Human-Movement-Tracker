```python
from ultralytics import YOLO

model = YOLO(r"runs\detect\train\weights\best.pt")
results = model(
    source=r"People-Detection-10\test\images\000077_jpg.rf.86614d2155ccf19053745e610adca4ab.jpg",
    show=True,      # Show image with detections (requires GUI support)
    conf=0.5,       # Confidence threshold
    save=True,      # Save predictions to runs/detect/predict
    save_txt=True,  # Save labels to .txt
    save_conf=True  # Save confidences with labels
)

import matplotlib.pyplot as plt

for r in results:
    plt.imshow(r.plot())  # r.plot() returns a numpy image with detections
    plt.axis("off")
    plt.show()
```
